```
import tensorflow as tf
from tensorflow import keras
from keras import backend as K
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.datasets import load_files
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
```

```
train_dir='./fruits1/Training/'
test_dir='./fruits1/Test/'
```

```
def load_file(path):
  data=load_files(path)
  files=np.array(data['filenames'])
  targets=np.array(data['target'])
  targets_name=np.array(data['target_names'])
  return files, targets, targets_name
```

```
x_train, y_train, target_labels=load_file(train_dir)
x_test, y_test,_=load_file(test_dir)
print('Training Example:', x_train.shape[0])
print('Test Example:', x_test.shape[0])
```

```
print("All Class:",np.amax(y_train)+1)
for i in range(len(target_labels)):
  print("Class "+str(i),target_labels[i])
```

```
import seaborn as sns
Y_train=tf.keras.utils.to_categorical(y_train, num_classes=np.amax(y_train)+1)
Y_test=tf.keras.utils.to_categorical(y_test, num_classes=np.amax(y_train)+1)
df=pd.DataFrame()
df["labels"]=y_train
lab=df['labels']
sns.countplot(lab)
```

```
X=[]
x_test1=[]
for i in range(len(x_train)):
  img=cv2.imread(x_train[i])
  X.append(img)
for i in range(len(x_test)):
  img=cv2.imread(x_test[i])
  x_test1.append(img)
X=np.asarray(X)
x_test1=np.asarray(x_test1)
```

```
X_train, x_val, y_trainhot, y_val=train_test_split(X, Y_train, test_size=0.2, random_state=320)
```

```
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```
plotImages(X_train[:5])
```

```
X_train=X_train/255
x_test1=x_test1/255
x_val=x_val/255
```

```
class mish(tf.keras.layers.Layer):
  def __init__(self):
    super().__init__()
  def call(self,x):
    return x*K.tanh(K.softplus(x))
```

```
my_callback=[tf.keras.callbacks.TensorBoard('./log')]
```

```
model = tf.keras.Sequential()
model.add(tf.keras.layers.Conv2D(filters = 16, kernel_size = 2, input_shape=(100,100,3),padding='same'))
model.add(mish())
model.add(tf.keras.layers.MaxPooling2D(pool_size=2))
model.add(tf.keras.layers.Conv2D(filters = 32,kernel_size = 2,padding='same'))
model.add(mish())
model.add(tf.keras.layers.MaxPooling2D(pool_size=2))

model.add(tf.keras.layers.Conv2D(filters = 64,kernel_size = 2,padding='same'))
model.add(mish())
model.add(tf.keras.layers.MaxPooling2D(pool_size=2))

model.add(tf.keras.layers.Conv2D(filters = 128,kernel_size = 2,padding='same'))
model.add(mish())
model.add(tf.keras.layers.MaxPooling2D(pool_size=2))

model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(150))
model.add(mish())
model.add(tf.keras.layers.Dropout(0.4))
model.add(tf.keras.layers.Dense(np.amax(y_train)+1,activation = 'softmax'))
model.summary()
```

```
model.compile(optimizer=tf.keras.optimizers.Adam(lr=1e-4), loss=tf.keras.losses.categorical_crossentropy, metrics=['acc'])
history=model.fit(X_train, y_trainhot, batch_size=32, epochs=20, callbacks=my_callback, validation_data=(x_val, y_val), verbose=1)
```

```
plt.plot(history.history['acc'], label='accuracy')
plt.plot(history.history['val_acc'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5,1])
plt.legend(loc='lower right')
```

```
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim([0.5,1])
plt.legend(loc='lower right')
```

```
model.evaluate(x_test1, Y_test, verbose=1)
```

```
tf.keras.models.save_model(model, filepath='./model')
```

```
from sklearn.metrics import confusion_matrix
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')


Y_pred = model.predict(x_test1)
Y_pred_classes = np.argmax(Y_pred,axis = 1) 
Y_true = np.argmax(Y_test,axis = 1) 
confusion_mtx = confusion_matrix(Y_true, Y_pred_classes) 
plot_confusion_matrix(confusion_mtx, classes = range(9)) 
```
